```python
from sklearn.model_selection import train_test_split
df_train,df_test=train_test_split(df,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
num=['instant', 'season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday',
       'weathersit', 'temp', 'atemp', 'hum', 'windspeed', 'casual',
       'registered', 'cnt']

scaler=MinMaxScaler()
# Convert df_train and df_test back to pandas DataFrames
df_train = pd.DataFrame(df_train, columns=df.columns)
df_test = pd.DataFrame(df_test, columns=df.columns)

# Now you can use column names for indexing
df_train[num]=scaler.fit_transform(df_train[num])
df_test[num]=scaler.transform(df_test[num])
```

```python
df_train
```

```python
y_train=df_train.pop('cnt')
x_train=df_train
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
def build_model(X,y):
    X = sm.add_constant(X) # Adding the constant
    lm = sm.OLS(y,X).fit() # fitting the model  sm.OLS.fit(y,x)
    print(lm.summary())    # model summary
    return lm
```

```python
model1=build_model(x_train,y_train)
```

```python
x=x_train.drop(['instant','yr', 'mnth', 'holiday', 'weekday', 'workingday', 'temp', 'atemp', 'windspeed' ],axis=1)
```

```python

lm = LinearRegression()
#lm.fit(X_train1,y_train)

rfe = RFE(lm,n_features_to_select=3)
rfe = rfe.fit(x, y_train)
```

```python
X_train_rfe = x[x.columns[rfe.support_]]
```

```python
X_train_rfe
```

```python
model2=build_model(X_train_rfe,y_train)
```

```python
X_train_new1=sm.add_constant(X_train_rfe)
y_train_pred=model2.predict(X_train_new1)
```

```python
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
y_test
```

```python
# Assuming X_train_rfe contains the selected features used for training model2
feature_names = X_train_rfe.columns.tolist()

# Add a constant term to X_test, similar to what was done for training
X_test_new = sm.add_constant(X_test[feature_names])

# Now predict using the modified X_test_new
y_pred = model2.predict(X_test_new)
```

```python
from sklearn.metrics import r2_score
print("Test Prediction R-Sqrd: ", r2_score(y_test, y_pred))
```
